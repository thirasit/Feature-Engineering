# **Mean or median Imputation**

Mean / median imputation consists in replacing missing values by the variable's mean or median.

Performimg mean and median imputation utilizing pandas, Scikit-learn and Feature-engine.

```python
pip install feature-engine
```

```python
import pandas as pd

# to split the data sets:
from sklearn.model_selection import train_test_split

# to impute missing data with sklearn:
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# to impute missing data with Feature-engine:
from feature_engine.imputation import MeanMedianImputer
```

## **Load data**

```python
data = pd.read_csv("credit_approval_uci.csv")
data.head()
```

## **Split data in train and test sets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop("target", axis=1),
    data["target"],
    test_size=0.3,
    random_state=0,
)

X_train.shape, X_test.shape
```

```python
# Let's inspect the proportion of missing
# values per variable:
X_train.isnull().mean()
```

## **Select the variables to impute**

```python
numeric_vars = X_train.select_dtypes(exclude="O").columns.to_list()
numeric_vars
```

## **Mean / median imputation with pandas**

[pd.fillna](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html)

```python
# Learn the variables median values:
median_values = X_train[numeric_vars].median().to_dict()
median_values
```

```python
# Replace missing data by the median:
X_train = X_train.fillna(value=median_values)
X_test = X_test.fillna(value=median_values)
```

```python
# Corroborate absence of missing values:
X_train[numeric_vars].isnull().sum()
```

```python
# Corroborate absence of missing values:
X_test[numeric_vars].isnull().sum()
```

## **Mean / median imputation with Scikit-learn**

[SimpleImputer](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html)

```python
# Split data into train and test set:

X_train, X_test, y_train, y_test = train_test_split(
    data.drop("target", axis=1),
    data["target"],
    test_size=0.3,
    random_state=0,
)
```

```python
numeric_vars
```

```python
# Make a list with the non-numerical variables:
remaining_vars = [var for var in X_train.columns if var not in numeric_vars]
remaining_vars
```

```python
# Set up the imputer to replace missing data
# with the median:
imputer = SimpleImputer(strategy="median")

# Indicate which variables to impute:
ct = ColumnTransformer(
    [("imputer", imputer, numeric_vars)], remainder="passthrough")

# Find the median value per variable:
ct.fit(X_train)
```

```python
# Check the medians that will be used in
# the imputation:
ct.named_transformers_.imputer.statistics_
```

```python
# Replace missing data:
X_train = ct.transform(X_train)
X_test = ct.transform(X_test)
X_train
```

```python
# Convert returned array to a pandas dataframe:
X_train = pd.DataFrame(
    X_train,
    columns=numeric_vars + remaining_vars,
)

X_train.head()
```

```python
# Corroborate absence of missing values:
X_train[numeric_vars].isnull().sum()
```

```python
# Convert returned array to a pandas dataframe:
X_test = pd.DataFrame(
    X_test,
    columns=numeric_vars + remaining_vars,
)

X_test.head()
```

```python
# Corroborate absence of missing values:
X_test[numeric_vars].isnull().sum()
```

## **Mean / Median imputation with Feature-engine**

[MeanMedianImputer](https://feature-engine.readthedocs.io/en/latest/api_doc/imputation/MeanMedianImputer.html)

```python
# Split data into train and test set:
X_train, X_test, y_train, y_test = train_test_split(
    data.drop("target", axis=1),
    data["target"],
    test_size=0.3,
    random_state=0,
)
```

```python
# Set up the imputer to replace missing data
# with the median:
imputer = MeanMedianImputer(
    imputation_method="median",
    variables=numeric_vars,
)

# Find the median values:
imputer.fit(X_train)
```

```python
# The median values per variable:
imputer.imputer_dict_
```

```python
# Replace missing data with the median:
X_train = imputer.transform(X_train)
X_test = imputer.transform(X_test)
```

```python
# Corroborate absence of missing values:
X_train[numeric_vars].isnull().sum()
```

```python
# Corroborate absence of missing values:
X_test[numeric_vars].isnull().sum()
```
